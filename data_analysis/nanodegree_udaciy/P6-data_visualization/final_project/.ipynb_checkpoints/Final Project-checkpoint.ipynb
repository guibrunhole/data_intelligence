Para usar o *Prophet* eu vou dividir o nosso dicionário principal (**my_file**) em n dicionários, sendo um para cada *local_site_name*. Nesse dicionário irei colocar somente três colunas:
- **date_local**
- **celsius**
- **local_site_name**
- **type:** será preenchido com *predicted*

Assim irei efetuar a predição para cada local e no final eu junto novamente tudo em um único dicionário :)

```python
local_name = my_file['local_site_name'].unique()
```

```python
len(local_name)
```

Temos **221** locais diferentes :)

Como temos dados desde 2000-01-01, irei realizar a predição para 3 meses :)

O único problema é que temos um *local_site_name* como **nan**. De início eu não sabia, aí tive vários erros. 
Tive que removê-lo :)

```python
local_name = np.delete(local_name, 33); # removing nan
```

```python
def predict (real_data, local_name):
    
    model = Prophet()
    model.fit(real_data)
    future = model.make_future_dataframe(periods=90) ## 90 dias
    forecast = model.predict(future)
    
    forecast['local_site_name'] = local_name
    forecast['type'] = 'predicted'
    
    return forecast
```
