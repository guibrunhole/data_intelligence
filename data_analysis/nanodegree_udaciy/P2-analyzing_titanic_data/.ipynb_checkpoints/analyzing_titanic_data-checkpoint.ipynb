### Uma breve introdução sobre o Titanic:

O <b>RMS Titanic</b> foi um navio de passageiros britânico operado pela White Star Line e construído pelos estaleiros da Harland and Wolff em Belfast. Foi a segunda embarcação da Classe Olympic de transatlânticos depois do RMS Olympic e seguido pelo HMHS Britannic. Projetado pelos engenheiros navais Alexander Carlisle e Thomas Andrews, sua construção começou em março de 1909 e ele foi lançado ao mar em maio de 1911. O Titanic foi pensado para ser o navio mais luxuoso e mais seguro de sua época, gerando lendas que era supostamente <b>"inafundável"</b>.

# 1- Limpando os dados

Vamos arrumar também os nomes, apenas para ficar mais fácil de lermos. Irei criar duas novas colunas, chamadas *FirstName* e *LastName* :)

Irei criar duas funções, uma para o *sobrenome* (last_name) e outra para o *primeiro nome* (first_name). 
Irei aplicá-las ao DataFrame utilizando o método *apply*, que já executa a função para todos os registros do meu DataFrame, sem que eu precise criar um loop por fora :)

```python
def first_name(input):
        return input.split(',')[1]
    
def last_name(input):
    return input.split(',')[0]
```

```python
titanic_data['FirstName'] = titanic_data['Name'].apply(first_name)
titanic_data['LastName'] = titanic_data['Name'].apply(last_name)
```

```python
titanic_data.head()
```

Além disso, para não termos informações duplicadas, irei excluir a coluna original, *Name*.

Pra realizarmos essa operação utilizamos a função *drop* já existente no Pandas. É bem simples utilizá-la, dê uma olhada:

```python
titanic_data = titanic_data.drop('Name', 1)

## verificando se realmente foi excluída corretamente
titanic_data.head()
```

Último tratamento que iremos fazer é atualizar a coluna *Pclass* para uma label de fato e não somente números. 
Na descrição do dataset temos que: (1 = 1st; 2 = 2nd; 3 = 3rd). Vamos, portanto, realizar essa substituição :)

```python
titanic_data['Pclass'] = titanic_data['Pclass'].replace(1,'Upper')
titanic_data['Pclass'] = titanic_data['Pclass'].replace(2,'Middle')
titanic_data['Pclass'] = titanic_data['Pclass'].replace(3,'Lower')
```

### Dando continuidade à análise acima, vamos ainda ver mais gráficos sobre o sexo das pessoas

Que tal começarmos analisando a proporção de homens e mulheres que embarcaram no navio?
O que precisamos para fazer isso é realizar uma contagem agrupando pelo sexo. Para a nossa sorte já existe a função *groupby* escrita no Pandas, assim fica bem simples. Dê uma olhada :D

```python
pie_plot = pd.Series(titanic_data['Sex'].groupby(titanic_data['Sex']).count())

pie_plot.plot(kind='pie', figsize=(4,4), autopct='%.2f')
```

Assim podemos ver que aprox. 2/3 da tripulação era do sexo masculino. Quando você precisa comparar a quantidade entre determinadas categorias a melhor forma de visualizar isso é através de um gráfico de pizza. Mas lembre-se, ele só é bom até uma certa quantidade de quebras, acredito que no máximo 4. Após isso já fica difícil de visualizar, aí já acredito que seja melhor utilizar um *Tree-map*. :)

### Vamos fazer a mesma comparação, porém com os sobreviventes e depois com os falecidos!

Realizei o mesmo código mostrado acima, porém troquei para utilizar o DataFrame *survived*

```python
pie_plot = pd.Series(survived['Sex'].groupby(survived['Sex']).count())

pie_plot.plot(kind='pie', figsize=(4,4), autopct='%.2f')
```

Bom, aqui já podemos ver que mesmo a grande maioria da tripulação sendo masculina, foram as mulheres que mais sobreviveram, cerca também de 2/3 dos sobreveviventes eram mulheres :)

Realizando a mesma análise com o conjunto de dados dos falecidos temos:

```python
pie_plot = pd.Series(killed['Sex'].groupby(killed['Sex']).count())

pie_plot.plot(kind='pie', figsize=(4,4), autopct='%.2f')
```

Agora podemos ter certeza absoluta de que os homens sobreviveram muito menos do que as mulheres.
